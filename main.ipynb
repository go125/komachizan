```python
import random
```

```python
def make_oplist():
    op_list=[]
    for i in range(0,9):
        rand_int=random.randint(0,3)
        if rand_int==0:
            op_list.append("+")
        if rand_int==1:
            op_list.append("-")
        if rand_int==2:
            op_list.append("*")
        if rand_int==3:
            op_list.append("/")
    return op_list
```

```python
tmp=make_oplist()
```

```python
def make_func(oplist):
    func_str=""
    for i in range(1,9):
        func_str=func_str+str(i)
        func_str=func_str+oplist[i-1]
    func_str=func_str+"9"
    return func_str
```

```python
make_func(tmp)
```

```python
eval(make_func(tmp))
```

```python
good_result=[]
```

```python
def main(max_count=1000000,margin=0.00001):
    global good_result
    count=0
    while count<=max_count:
        if count%1000==0:
            print(str(count)+"/"+str(max_count)+"has finished.")
        oplist=make_oplist()
        func=make_func(oplist)
        res=eval(func)
        error=res-100
        if abs(error)<margin:
            print(func)
            good_result.append(func)
        count+=1
```

```python
main()
```
