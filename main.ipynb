```python
def make_oplist(num):
    op_list=[]
    for i in range(0,9):
        rand_int=num%17
        if rand_int==0:
            op_list.append("+")
        if rand_int==1:
            op_list.append("-")
        if rand_int==2:
            op_list.append("*")
        if rand_int==3:
            op_list.append("/")
        if rand_int==4:
            op_list.append(")+")
        if rand_int==5:
            op_list.append(")-")
        if rand_int==6:
            op_list.append(")*")
        if rand_int==7:
            op_list.append(")/")
        if rand_int==8:
            op_list.append("+(")
        if rand_int==9:
            op_list.append("-(")
        if rand_int==10:
            op_list.append("*(")
        if rand_int==11:
            op_list.append("/(")
        if rand_int==12:
            op_list.append(")+(")
        if rand_int==13:
            op_list.append(")-(")
        if rand_int==14:
            op_list.append(")*(")
        if rand_int==15:
            op_list.append(")/(")
        if rand_int==16:
            op_list.append("")
        num=num//17
    return op_list
```

```python
def make_func(oplist):
    func_str=""
    for i in range(1,9):
        func_str=func_str+str(i)
        func_str=func_str+oplist[i-1]
    func_str=func_str+"9"
        
    return func_str
```

```python
good_result=[]
```

```python
def main(max_count=17**8,margin=0.00001):
    global good_result
    count=0
    while count<=max_count:
        if count%1000000==0:
            print(str(count)+"/"+str(max_count)+" has finished.")
        oplist=make_oplist(count)
        func=make_func(oplist)
        try:
            eval(func)
        except SyntaxError:
            count+=1
            continue
        except ZeroDivisionError:
            count+=1
            continue
        res=eval(func)
        error=res-100
        if abs(error)<margin:
            print(func)
            good_result.append(func)
        count+=1
```

```python
main()
```
